```python
train_data["Open Date"] = pd.to_datetime(train_data["Open Date"])
train_data["Years"] = train_data["Open Date"].dt.year
train_data.drop("Open Date",axis=1,inplace=True)

test_data["Open Date"] = pd.to_datetime(test_data["Open Date"])
test_data["Years"] = test_data["Open Date"].dt.year
test_data.drop("Open Date",axis=1,inplace=True)
```

```python
train_data.drop("City",axis=1,inplace=True)
test_data.drop("City",axis=1,inplace=True)
train_data['City Group'] = train_data['City Group'].replace({'Big Cities': 0, 'Other': 1})
test_data['City Group'] = test_data['City Group'].replace({'Big Cities': 0, 'Other': 1})
train_data['Type'] = train_data['Type'].replace({'FC': 0, 'IL': 1,'DT': 2})
test_data['Type'] = test_data['Type'].replace({'FC': 0, 'IL': 1,'DT': 2})
```

```python
test_data['Type'].value_counts()
```

```python
removed_indices = test_data[test_data['Type'] == 'MB'].index
```

```python
test_data = test_data.loc[test_data['Type'] != 'MB']
```

```python
final_data = final_data.drop(removed_indices)
```

```python
X_train = train_data.drop(['Id', 'revenue'], axis=1)  # Assuming 'Id' and 'revenue' are not features
y_train = train_data['revenue']
```

```python
X_test = test_data.drop('Id', axis=1)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred_linear = linear_reg.predict(X_test)
y_pred_linear
```

### mean squared error for Linear Regression

```python
from sklearn.metrics import mean_squared_error
y_pred_linear = linear_reg.predict(X_test)
true_values_test = final_data['Id']
mse_linear = mean_squared_error(true_values_test, y_pred_linear)
```

```python
mse_linear
```

## accuracy

```python
errors_linear = abs(y_pred_linear - true_values_test)
mape_linear = 100 * (errors_linear / true_values_test)
accuracy_linear = 100 - np.mean(mape_linear)
print('Linear Regression Accuracy:', round(accuracy_linear, 2), '%.')
```

```python
feature_importances
```

```python
# Created a DataFrame to store feature importances
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importances})

#  Sorting the DataFrame by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
feature_importance_df
```
